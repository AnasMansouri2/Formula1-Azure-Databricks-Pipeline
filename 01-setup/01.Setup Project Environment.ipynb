# Set-up the project environment for F1 Project
1. Create External Location databricks-course-ext-dl77-formula1
2. Create Catalog Formula 1
3. Create Schemas landing , bronze , silver and gold.
4. Create Volume Files in landing schema.

**ACCESS CLOUD STORAGE**

```sql
%fs ls 'abfss://formula1@databrickscourseextdl77.dfs.core.windows.net/landing'
```

**CREATE EXTERNAL LOCATION**

```sql
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricks_course_ext_dl77_formula1
URL 'abfss://formula1@databrickscourseextdl77.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-course-sc`) -- Changé ' par `
COMMENT 'External location for the formula1 container';
```

**Create Formula 1 Catalog**

```sql
show catalogs
```

```sql
CREATE CATALOG IF NOT EXISTS formula1
    MANAGED LOCATION 'abfss://formula1@databrickscourseextdl77.dfs.core.windows.net/'
    COMMENT 'This is the main catalog for the formula 1 project ';
```

**Create Schemas : landing , bronze , silver and gold.**

```sql
CREATE SCHEMA IF NOT EXISTS formula1.landing;
CREATE SCHEMA IF NOT EXISTS formula1.bronze
    MANAGED LOCATION 'abfss://formula1@databrickscourseextdl77.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS formula1.silver
    MANAGED LOCATION 'abfss://formula1@databrickscourseextdl77.dfs.core.windows.net/silver';    
CREATE SCHEMA IF NOT EXISTS formula1.gold
    MANAGED LOCATION 'abfss://formula1@databrickscourseextdl77.dfs.core.windows.net/gold';
```

```sql
USE CATALOG formula1
```

```sql
SHOW SCHEMAS
```

**Create Volume Files in landing schema**

```sql
CREATE EXTERNAL VOLUME formula1.landing.files
LOCATION 'abfss://formula1@databrickscourseextdl77.dfs.core.windows.net/landing'
```
